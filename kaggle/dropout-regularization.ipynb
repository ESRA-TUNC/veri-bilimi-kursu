```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/kaggle/input/connectionist-bench-sonar-mines-vs-rocks/sonar.all-data.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df['Label'].value_counts()
```

```python
X = df.drop(['Label'],axis='columns')
y = df['Label']
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # R=True M=False
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

## Create the DeepLearning Model and See the Results

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
model.predict(X_test).reshape(-1)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[10])
# round 

y_pred = np.round(y_pred)
print(y_pred[10])
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## Apply the Drop Out Layer and Check the Result

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
     keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
